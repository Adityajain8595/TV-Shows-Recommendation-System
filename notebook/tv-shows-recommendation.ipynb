```python
shows = pd.read_csv('TMDB_tv_dataset_v3.csv')
shows.head()
```

```python
shows = shows[['id', 'name', 'overview', 'genres', 'poster_path', 'type', 'tagline', 'created_by', 'networks', 
               'number_of_seasons', 'number_of_episodes', 'vote_count', 'vote_average']]
shows.head()
```

```python
shows = shows.dropna(subset=['name', 'overview', 'genres', 'poster_path'])
shows = shows[shows['type'].isin(['Scripted', 'Documentary', 'Miniseries'])]
shows.drop_duplicates(inplace=True)
shows = shows.reset_index(drop=True)

shows.info()
```

```python
shows['overview'] = shows['overview'].fillna('')
shows['tagline'] = shows['tagline'].fillna('')

shows['genres'] = shows['genres'].apply(
    lambda x: x.split(', ') if isinstance(x, str) and x.strip() else []
)

shows['type'] = shows['type'].apply(
    lambda x: x.split(', ') if isinstance(x, str) and x.strip() else []
)

shows['created_by'] = shows['created_by'].apply(
    lambda x: x.split(', ') if isinstance(x, str) and x.strip() else []
)

shows['networks'] = shows['networks'].apply(
    lambda x: x.split(', ') if isinstance(x, str) and x.strip() else []
)

shows.head()
```

```python
shows['tags'] = (
    shows['name'] + ' ' + 
    shows['overview'] + ' ' + shows['tagline'] + ' ' + 
    (shows['genres'].apply(lambda x: ' '.join(x)) + ' ')*2 + 
    shows['type'].apply(lambda x: ' '.join(x)) + ' ' + 
    (shows['created_by'].apply(lambda x: ' '.join(x)) + ' ')*3 + 
    shows['networks'].apply(lambda x: ' '.join(x))
)

shows.head()
```

```python
import re
from bs4 import BeautifulSoup
```

```python
# Remove url
shows['tags'] = shows['tags'].apply(lambda x: re.sub(r'(http|https|ftp|ssh)://([\w_-]+(?:(?:\.[\w_-]+)+))([\w.,@?^=%&:/~+#-]*[\w@?^=%&/~+#-])?', '' , str(x)))

# Remove html tags
shows['tags'] = shows['tags'].apply(lambda x: BeautifulSoup(x, 'lxml').get_text())
```

```python
import os
import pickle
from dotenv import load_dotenv
load_dotenv()

embeddings_path = os.getenv("EMBEDDINGS_PATH")
model_path = os.getenv("MODEL_PATH")

with open(embeddings_path, "rb") as f:
    embeddings = pickle.load(f)

with open(model_path, "rb") as f:
    model = pickle.load(f)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
from langchain_groq import ChatGroq
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import PydanticOutputParser
from pydantic import BaseModel, Field

# Define the Schema
class Explanation(BaseModel):
    explanation: str = Field(..., description="A reason for why a recommended show is a good match.")

parser = PydanticOutputParser(pydantic_object=Explanation)

os.environ['GROQ_API_KEY'] = os.getenv('GROQ_API_KEY')
model = ChatGroq(model="llama-3.3-70b-versatile")

# Explanation function
def get_explanation(target_row, rec_row):
    template = """
    You are a razor-sharp TV critic known for your witty, insightful observations, 
    and explaining why a fan of "{target_name}" would love "{rec_name}".
        
        ---
        METADATA FOR ANALYSIS:
        TARGET SHOW: {target_name}
        - Genres: {target_genres}
        - Creator/Team: {target_creator}
        - Synopsis: {target_overview}
        
        RECOMMENDED SHOW: {rec_name}
        - Genres: {rec_genres}
        - Creator/Team: {rec_creator}
        - Synopsis: {rec_overview}
        ---
        
        INSTRUCTIONS:
        1. IDENTIFY THE HOOK by looking for shared universes, shared creators, or strong thematic parallels.
        2. USE EXTERNAL LORE of these shows' plots, character arcs, and cultural impact.
        3. Be SPECIFIC about plot elements, character traits, or storytelling techniques
        4. Use ACTIVE, vivid language (not "this show has" or "features")
        5. NO generic phrases like "captivating storyline", "must-watch", "rollercoaster"
        6. NO starting with "Fans of X will enjoy..." or "If you liked X..."
        7. Be CONCISE and sound like a knowledgeable friend, not a marketing robot
        
        Write ONE sharp, specific explanation that makes me instantly want to watch {rec_name}.
        
        {format_instructions}
        
        JSON RESPONSE:"""
    
    prompts = PromptTemplate(
        input_variables=["target_name", "target_overview", "target_genres", "target_creator", "rec_name", "rec_overview", "rec_genres", 
                         "target_creator"],
        template=template,
        partial_variables={"format_instructions": parser.get_format_instructions()}
    )
    
    chain = prompts | model | parser
    
    response = chain.invoke({
        "target_name": target_row['name'],
        "target_overview": target_row['overview'],
        "target_genres": ', '.join(target_row['genres']),
        "target_creator": ', '.join(target_row['created_by']),
        "rec_name": rec_row['name'],
        "rec_overview": rec_row['overview'],
        "rec_genres": ', '.join(rec_row['genres']),
        "rec_creator": ', '.join(rec_row['created_by'])
    })
    
    return response.explanation

def recommend(show_name, vote_count=0, vote_average=0, seasons=10, episodes=100):
    match = shows[shows['name'].str.lower() == show_name.lower()]
    
    if match.empty:
        print("Show not found")
        return
    
    idx = match.index[0]
    target_row = shows.iloc[idx]
    target_embedding = embeddings[idx].reshape(1, -1)
    
    filtered_indices = shows[
        (shows['vote_count'] >= vote_count) & 
        (shows['vote_average'] >= vote_average) &
        (shows['number_of_seasons'] <= seasons) &
        (shows['number_of_episodes'] <= episodes) &
        (shows.index != idx)
    ].index
    
    filtered_embeddings = embeddings[filtered_indices]
    similarity_scores = cosine_similarity(target_embedding, filtered_embeddings)[0]
    
    results = sorted(zip(filtered_indices, similarity_scores), key=lambda x: x[1], reverse=True)[:5]
    
    print(f"Recommendations for '{show_name}':\n" + "-"*30)
    
    for i, (f_idx, score) in enumerate(results):
        rec_row = shows.iloc[f_idx] 
        print(f"{i+1}. {rec_row['name']} -> (Score: {score:.4f})")
        
        explanation = get_explanation(target_row, rec_row)
        print(f"    💡 Why: {explanation}\n")
        
recommend("Better Call Saul", vote_average=8)
```
